# List Comprehensions

In addition to sequence operations and list methods, Python includes a more advanced operation called a list comprehension.

List comprehensions allow us to build out lists using a different notation. You can think of it as essentially a one line <code>for</code> loop built inside of brackets. For a simple example:
## Example 1

```python
mystr = 'hello'
```

```python
mylist = []

for letter in mystr:
    mylist.append(letter)
```

```python
mylist
```

```python
#[item for item in the iterate list or object]
#if you want to form operation form it in the first item
mylist1 = [letter for letter in 'hello_world']
```

```python
mylist1
```

```python
# Grab every letter in string
lst = [x for x in 'word']
```

```python
# Check
lst
```

```python
num = [n for n in range(5)]
```

```python
num
```

```python
mylist3 = [x for x in range(0,11)]
mylist3
```

This is the basic idea of a list comprehension. If you're familiar with mathematical notation this format should feel familiar for example: x^2 : x in { 0,1,2...10 } 

Let's see a few more examples of list comprehensions in Python:
## Example 2

```python
# Square numbers in range and turn into list >> form operations in the first item
lst = [x**2 for x in range(0,11)]
```

```python
lst
```

## Example 3
Let's see how to add in <code>if</code> statements:

```python
# Check for even numbers in a range 
lst = [x for x in range(11) if x % 2 == 0]
```

```python
lst
```

```python
even = [x if x%2 == 0 else 'ODD' for x in range (0,14)]
```

```python
even
```

```python
ll = [xx  if xx%2==0 else 'This is an ODD number!!' for xx in range(11)]
```

```python
print(ll)
```

## Example 4
Can also do more complicated arithmetic:

```python
# Convert Celsius list to Fahrenheit list
celsius = [0,10,20.1,34.5]

fahrenheit = [((9/5)*temp + 32) for temp in celsius ]

fahrenheit
```

## Example 5
We can also perform nested list comprehensions, for example:

```python
mylist4 = []

for x in [2,4,6]:
    for y in [100,200,300]:
        mylist4.append(x*y)
```

```python
mylist4
```

```python
list5 = [x*y for x in [2,4,6] for y in [100,200,300]]
```

```python
list5
```

```python
lst = [ x**2 for x in [x**2 for x in range(11)]]
lst
```
